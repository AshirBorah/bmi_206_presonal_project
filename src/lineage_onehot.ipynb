```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd
```

```python
sample_info = pd.read_csv('../data/raw/sample_info.csv')
```

```python
sample_info.head()
```

```python
enc = OneHotEncoder(handle_unknown='ignore')
enc = enc.fit(sample_info.loc[:,['lineage', 'lineage_subtype', 'lineage_sub_subtype', ]])
onehot_lineages = enc.fit_transform(sample_info.loc[:,['lineage', 'lineage_subtype', 'lineage_sub_subtype', ]])
```

```python
onehot = pd.DataFrame.sparse.from_spmatrix(onehot_lineages, columns = enc.get_feature_names_out(), index=sample_info.DepMap_ID)
onehot = onehot.astype('int32')
onehot.head()
```

```python
onehot.to_csv('../data/raw/onehot_lineage.csv')
```
